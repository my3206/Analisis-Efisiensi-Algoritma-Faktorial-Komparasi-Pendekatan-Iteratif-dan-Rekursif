```python
import time
import sys
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Biar rekursif aman sampai n tertentu
sys.setrecursionlimit(10000)
```

```python
# ALGORITMA FAKTORIAL
def factorial_iterative(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

def factorial_recursive(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial_recursive(n - 1)
```

```python
# FUNGSI UKUR WAKTU
def measure_time(func, n):
    start = time.time()
    func(n)
    end = time.time()
    return end - start
```

```python
# PENGUJIAN
input_sizes = [10, 100, 500, 1000, 5000]

iter_times = []
recur_times = []

for n in input_sizes:
    iter_times.append(measure_time(factorial_iterative, n))
    recur_times.append(measure_time(factorial_recursive, n))
```

```python
# TABEL PERBANDINGAN
df = pd.DataFrame({
    "n": input_sizes,
    "Waktu Rekursif (s)": recur_times,
    "Waktu Iteratif (s)": iter_times
})
df
```

```python
# GRAFIK PERBANDINGAN
plt.figure()
plt.plot(input_sizes, iter_times, marker='o', label='Iteratif')
plt.plot(input_sizes, recur_times, marker='o', label='Rekursif')

plt.xlabel("Ukuran Input (n)")
plt.ylabel("Running Time (s)")
plt.legend()
plt.grid(True)
plt.show()
```
